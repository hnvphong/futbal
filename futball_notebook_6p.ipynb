#### Phòng thủ:
Tên hành động kèm chú giải, trong các video hoặc trao đổi khi tập luyện/thi đấu, tôi sẽ chỉ dùng tên quy tắc cho ngắn gọn. Ví dụ: khi tôi hét lên 1 1, đoạt lại, phá bóng, giữ bóng lại..., các cầu thủ sẽ biết cần phải làm gì.
- 1 kèm 1: trong trường hợp cầu thủ phòng ngự đội bạn không tham gia tấn công, không cần ở lại phía sân đội bạn,<b> vậy trong tình huống đội bạn tràn lên tấn công số đông (4, 5 người) -> áp dụng 1-1 -> cầu thủ tấn công của ta lui về hết</b>
- Cướp lại: ai mất bóng phải đuổi tranh chấp lại cho bằng được, nếu bóng đã bị chuyền đi-> vẫn theo kèm đối thủ vừa chuyền đi
- Theo người: khi bị đối phương chạy qua mặt, xử lí tương tự cướp lại
- Giữ khoảng cách: bóng đã tiến về phần sân nhà -> các cầu thủ tấn công lui về (KHÔNG ĐỨNG ĐỢI Ở PHÍA TRÊN)
- Phá bóng: khi không biết phải làm gì, hãy đưa bóng ra xa khung thành đội mình; hết biên dọc (bị đội bạn đá biên); hết biên ngang (bị đội bạn đá phạt góc)
- Giữ bóng lại: ai gần với bóng nhất giữa trái bóng lại chờ đồng đội lui về hỗ trợ, không cho đội bạn đá biên/đá phạt góc ngay
- <b>Tản ra: không bu hết người vào bóng (chỉ cần 1 người tranh chấp, đồng đội xung quanh chớp thời cơ bóng loạt khỏi vòng tranh chấp)</b>

- Khi đoạt được lại bóng -> chuyền đồng đội ở gần (nhìn thấy/nghe tiếng gọi) hoặc phá bóng
- Khi đồng đội đoạt được bóng -> các cầu thủ khác ngay lập tức di chuyển đến vị trí có thể nhận bóng và gọi (la/hét/hát) để đồng đội biết mà chuyền đến

- Giữ khoảng cách: các cạnh của tam giác tấn công không quá gần, quá xa (thường bằng khoảng cách hay tập luyện chuyền)

- Bóng ở cao trên phần sân đội bạn: 2 cầu thủ phòng ngự dâng cao, không để khoảng trống lớn sau lưng cầu thủ tấn công của đội mình, nhưng vẫn đảm bảo đúng vai trò của mình
- Kêu gọi: gọi to tên đồng đội <- <b>trong thi đấu sẽ bị áp sát và thể lực bị tiêu hao, việc có đủ thời gian và tỉnh táo để đưa ra quyết định chuyền đi đâu/cho ai/làm gì rất khó khăn. Nếu nhận thấy đồng đội đang có bóng và bản thân đang ở vị trí thuận lợi</b>
